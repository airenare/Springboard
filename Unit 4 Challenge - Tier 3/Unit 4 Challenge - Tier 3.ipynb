```python
from datetime import datetime


def get_start_end(year_of_int=None, for_20=False):
    '''Gets the start and end date of the observations. \
Set the most recent year of interest as year_of_int=2015 \
or None by default to end with the last date available. \
Set `for_20=True` to get only the last 20 years of \
observations, counting back from the set year_of_int'''
    
    # Setting the end_date
    if not year_of_int:
        end_date = boroughs_price['date'].max()
    else:
        date_of_int = pd.Timestamp(f'{year_of_int}-12-31 23:59:59', tz=None)
        end_date = boroughs_price[boroughs_price['date'] <= date_of_int]['date'].max()
        
    # Setting the start_date
    if for_20:
        # Getting the start of the year that was 20 years before last observation's year
        end_minus_20 = pd.Timestamp(f'{end_date.year - 20}-1-1 00:00:00', tz=None)
        start_date = boroughs_price[boroughs_price['date'] >= end_minus_20]['date'].min()
    else:
        start_date = boroughs_price['date'].min()
    
    return {'start_date': start_date, 'end_date': end_date}
    

# Set the get_start_end() arguments
# get_start_end.__doc__
year_of_interest = 2018
for_20y = True

# Getting the start and end dates
start_end = get_start_end(year_of_int=year_of_interest, for_20=for_20y)
start_date = start_end['start_date']
end_date = start_end['end_date']


# Subsetting the first and last observations for each borough
boroughs_price_diff = boroughs_price[boroughs_price['date'].isin([start_date, end_date])]

# Pivoting table to find price diff for each borough
boroughs_price_diff = boroughs_price_diff.set_index('borough').pivot_table(values=['avg_price'], 
                                     index='borough',
                                     aggfunc='diff').rename(columns={'avg_price':'price_diff'})
# Sorting descending
boroughs_price_diff = boroughs_price_diff.sort_values('price_diff', ascending=False)

# Plotting price difference
boroughs_price_diff.plot(kind='barh',
                        title=f'Average house price increase for London boroughs\nfrom {start_date.year} to {end_date.year}',
                         figsize=(8,8)
                        )
# Removing legend because there is only one variable
plt.legend().remove()

plt.xlabel("Price difference absolute, GBP")
plt.ylabel("")

plt.show()


```

**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:

```python
import numpy as np

# def price_ratio(df, borough, end_year:int, ratio=True):
def price_ratio_20(df, borough, end_year):
    # Transposing table so that Boroughs are rows and resetting index
    df = df.transpose().reset_index()
    
    # Setting column names and resetting index
    df.columns = df.iloc[0]
    df = df.drop(df.index[0]).reset_index(drop=True)
    df.rename(columns={"Unnamed: 0": "borough"}, inplace=True)

    # Removing NaT column
    df = df.drop(df.columns[df.columns.isna()], axis=1)

    # Selecting only rows with 32 London boroughs
    df = df.iloc[1:33].reset_index(drop=True)
    
    # Melting the table into tall format
    df = df.melt(id_vars=['borough'], value_vars=df.columns, var_name="date", value_name="avg_price")
    
    # Changing dtype of 'avg_price'
    df['avg_price'] = df['avg_price'].astype('float')
    df['borough'] = df['borough'].astype('string')
    
    # Subsetting for the borough
    df = df[df['borough']==borough]
    
    # Setting end_date
    date_of_int = pd.Timestamp(f'{end_year}-12-31 23:59:59', tz=None)
    end_date = df[df['date'] <= date_of_int]['date'].max()
    
    # Getting the start of the year that was 20 years before last observation's year
    end_minus_20 = pd.Timestamp(f'{end_date.year - 20}-1-1 00:00:00', tz=None)
    start_date = df[df['date'] >= end_minus_20]['date'].min()

    
    # Subsetting the first and last observations for each borough
    df = df[df['date'].isin([start_date, end_date])]
    
    # Calculating the return of assets (ROA) ratio
    start_price = df.iloc[0]['avg_price']
    end_price = df.iloc[1]['avg_price']
    df['roa'] = end_price/start_price
    df = df.reset_index(drop=True)
    df = df.loc[0, ["borough", "roa"]]
    
    return df


# Downloading and importing the table as DataFrame
url_LondonHousePrices = "https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls"
properties = pd.read_excel(url_LondonHousePrices, sheet_name='Average price', index_col= None)

# Initiating the resulting DataFrame
ratio_df = pd.DataFrame()

# Running a loop with the function for all boroughs in the table
last_year_of_interest = 2018
for borough in properties.columns[2:34]:
    result_df = price_ratio_20(properties, borough, last_year_of_interest)
    ratio_df = pd.concat([ratio_df, result_df], ignore_index=True, axis=1)

# Transposing the table
ratio_df = ratio_df.transpose()
# Changing dtypes
ratio_df['roa'] = ratio_df['roa'].astype('float')
ratio_df['borough'] = ratio_df['borough'].astype('string')
    
# Getting the borough with the largest price increase (return of assets) ratio over the course of 20 years since 1998 to 2018
greatest_growth_borough = ratio_df.nlargest(1, 'roa')

# Printing the result
result_borough = greatest_growth_borough['borough'].values[0]
result_increase = round(greatest_growth_borough['roa'].values[0], 1)

print(f'====[ RESULT ]====\n' \
      f'The {result_borough} borough of London has seen the {result_increase}-fold average \nprice increase (return of assets ratio) over the course of 20 years \nfrom {last_year_of_interest - 20} to {last_year_of_interest}, which is the greatest among all other London boroughs.'
)
```

```python
# Plotting the return of assets ratio for all boroughs
ratio_df.sort_values('roa', ascending=False).plot(kind='barh', 
                                                          x='borough', y='roa',
                                                         title = f"Price increase ratio for\
all London boroughs from {last_year_of_interest - 20} to {last_year_of_interest}", figsize=(8,8))
# Removing legend because there is only one variable
plt.legend().remove()

plt.xlabel("Price increase (ROA) ratio, times")
plt.ylabel("")

plt.show()
```
